#**Getting Started**

###Import the Seaborn library.

```python
import seaborn as sns 
```

###Load the sample .csv file included with Seaborn as a pandas dataframe named "categorical_data".

```python
categorical_data = sns.load_dataset('tips')
```

###Note that this dataframe has columns for the monetary value of each tip, the day each tip was given, and the sex of the person who gave the tip.

```python
categorical_data.head()
```

#**Plotting with Seaborn**

##**Bar Plots**


###Create a bar plot from the dataframe showing the mean tip on each day.

```python
sns.barplot(data=categorical_data, x='day', y='tip')
```

###Create a double-bar plot like above but where mean tips are seperated by sex. 
Notice that a key is automatically generated.


```python
sns.barplot(data=categorical_data, x='day', y='tip',hue='sex')
```

###In both of these bar graphs, the bars represent sample mean and the error bars represent the 95% confidence interval of each mean. However, you can manipulate these defaults using an additional python library, NumPy.

image.png

###Import NumPy.

```python
import numpy as np
```

###Create a bar graph like above where each bar represents the sample median instead of sample mean.

```python
sns.barplot(x='day', y='tip', data=categorical_data, estimator=np.median)
```

###Notice how the error bars have changed too. They now represent the 95% confidence interval of each sample median instead of the sample mean.

image.png

###Create a new bar graph where the error bars will now represent a 50% confidence interval.

```python
 sns.barplot(data=categorical_data, x='day', y='tip', ci=50)
```

###Create a new bar graph where the error bars now represent standard deviation (sd).
image.png 

```python
sns.barplot(data=categorical_data, x='day', y='tip', ci='sd')
```

###Create a bar graph with error bar whiskers that are 20% the width of each 

---

category. 

---



```python
sns.barplot(data=categorical_data, x='day', y='tip',capsize=0.2)
```

Create a bar graph with error bar whiskers that are 100% the width of each category.

```python
sns.barplot(data=categorical_data, x='day', y='tip',capsize=1)
```

##**Box Plots**

Let's review how to read box plots.

image.png

###Create a box plot showing the median tip on each day. 

```python
sns.boxplot(data=categorical_data, x='day', y='tip')
```

###Create a box plot like above, but remove the outlier points.

```python
sns.boxplot(data=categorical_data, x='day', y='tip',fliersize=0)
```

###Create a double-box plot like above but where mean tips are seperated by sex.
Notice that like before, a key is automatically generated.


```python
sns.boxplot(data=categorical_data, x='day', y='tip',fliersize=0,hue='sex') 
```

##**Strip Plots**

Why use strip plots? Consider the situation below:

image.png

Notice how the box plot does not describe the data's distribution (bi-modal) nor its sample size. 

###Create a strip plot showing all the tips on each day. 

```python
sns.stripplot(data=categorical_data, x='day', y='tip')
```

###Notice how the positions of datapoints are randomized each time the code is run.

```python
sns.stripplot(data=categorical_data, x='day', y='tip')
```

###Create a strip plot like above, but change the points from circles to triangles and adjust their size.

```python
sns.stripplot(data=categorical_data, x='day', y='tip',marker='^', size=5)
```

###Create a strip plot like the first one, but with the points more spread out.

By default, jitter=1

```python
sns.stripplot(data=categorical_data, x='day', y='tip',jitter=1)
```

###Create a strip plot like above, but reduce the alpha value (opacity) of each datapoint.
By default, alpha=1

```python
sns.stripplot(data=categorical_data, x='day', y='tip',jitter=.3,alpha=1)
```

###Create a double-strip plot where datapoints are labeled by color based on the customer's sex. 

```python
sns.stripplot(data=categorical_data, x='day', y='tip',jitter=.3,alpha=.6,hue='sex')
```

###Create a double-strip plot like above, but where alpha is reset and points of different color are seperated.
By default, dodge=False

```python
sns.stripplot(data=categorical_data, x='day', y='tip',hue='sex',dodge=True)
```
