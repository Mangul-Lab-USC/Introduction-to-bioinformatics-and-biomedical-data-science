After wrangling tabular data with Pandas and producing figures with Seaborn, you may want to share your exported DataFrames and figures with colleagues. 

The most transparent and organized way to do so is through GitHub, a cloud-based version control platform that will host your entire project as a repository, or repo for short.

Therefore, instead of jumping to GitHub, let's first learn the fundamentals of Git.

#**Introduction to Git subcommands**
Git is a UNIX command-line program; and by default, Google Colab notebooks run all code within a Python 3 runtime environment. Therefore, to use UNIX programs instead of Python code, users need to specify how they want the contents of a code block run. 

One way this can be achived is by starting each cell block with `%%bash` .
Here, the name Bash refers to a type of Unix shell. 

##**Run the `git` command in a Bash environment.**

```python
%%bash
git
```

Notice that by itself, the `git` command only prints information about its subcommands. These commands are what we will use to create our first repository.

To start, let's clear out the default current directory, `/content`,  that Google Colab puts us into every time we open a notebook. 

##**Run commands to prepare Google Colab for usage with Git.**

###**Remove the default `sample_data` folder from the `/content` directory.**

Here, we use the `rm` (remove) command with the `-r` (recursive) option.

```python
%%bash
rm -r sample_data
```

Additionally, we must create a special file called `.gitignore` to prevent the default Google Colab `.config` folder from interfering with our Git repository.




###**Create a `.gitignore` file that tells Git to ignore the hidden `.config` folder.**

```python
%%bash
cat > .gitignore
.*
.gitignore
```

Finally, before using Git we must specify who we are. This is to ensure that when multiple people are making changes to a Git project, everyone knows who made what changes. 

##**Set an identity with `git config`.**

This is typically performed exactly as follows, with an email or name written within the double-quotation marks. 


For the sake of this tutorial, we will use a dummy email and name. 

```python
%%bash
git config --global user.email "you@example.com"
git config --global user.name "Your Name"
```

Now we're ready to convert the current directory into a Git repository.

##**Convert `/content` into a Git repository with `git init`.**

```python
%%bash
git init
```

In addition to the message output above, you can confirm that the current directory is a Git repository by running the `ls` (list) command with the `-a` (all) option to display eveything within a directory, including hidden folders like `.git`.

```python
%%bash
ls -a
```

The presence of the `.git` folder confirms that we are inside a Git repository. From here, you can visualize the current version of the repository as a node like below.

image.png

Then, as we make changes to this repository, its history will be recorded as a series of nodes.

image.png

Note that each of the above nodes represents a distinct version of the repository. This is because Git does not continually save every little change to your repository. Instead changes must be saved (aka. commmited) to produce distinct versions. 

You may have done something similar before with Microsoft Word documents named as follows:

*   Manuscript.docx
*   Manuscript2.docx
*   Manuscript3.docx
*   ManuscriptFinal.docx
*   ManuscriptFinal_Final.docx
*   ManuscriptFinal_Final_Final.docx

However, instead of filling up your folder with a slew of document versions with potentially ambiguous names, Git will only store the latest version of the file while enabling you to restore previous versions should you need to. 

To demonstrate this, let us make changes to Git repository we have just made.

##**Make arbritrary changes to the Git repository with various Unix commands.**

The command below will produce a text file.

```python
%%bash
cat > textFile.txt
This is a text file.
It has the .fasta extension.
```

Now that we are satisfied with our changes let's save our progress as a new version. To do so, we use the `git add` and `git commit` commands.

`git add` will be used first to add all new files and/or subdirectories to a temporary staging area where changes can be reviewed. Then, once you are satisfied with all the changes, you can commit to the changes with `git commit`, creating a new version of your repository. 

image.png

##**Add one or more files to the staging area with `git add`.**

Here the `-A`/`--all`/`--no-ignore-removal` option is used to track all new, modified, and deleted files. 

```python
%%bash
git add -A
```

You can confirm that a new file was added to the staging area with the command below.

```python
%%bash
git diff --name-only --cached
```

##**Commit changes in the staging area with `git commit`.**

In addition to saving a new version of your repository, it is also standard practice to label your new version with a short message describing what has changed. 

This is performed with the `-m` (message) option followed by a description bounded by quotation marks. 

```python
%%bash
git commit -m "Added a text file."
```

Congratulations, you have made your first Git commit and are ready to make more!

We can do so by following the steps above. For example, if we check the file above, it contained an error.

```python
%%bash
cat textFile.txt
```

Instead of ending in .fasta, our text file ends in .txt . We can overwrite our file as follows:

```python
%%bash
cat > textFile.txt
This is a text file.
It has the .txt extension.
```

Next, we can save our changes by using `git add` and `git commit` again.

```python
%%bash
git add -A
git commit -m "Fixed the extension error in textFile.txt."
```

#**Following GitHub secruity protocol with SSH keys**

At the moment, your GitHub account is like a locked door without any keyholes.

image.png

Currently, there is no easy way to access what's inside.

Therefore, what we will explain here is how to add a keyhole to your GitHub account and produce a key that can unlock it. 

image.png

One way to achieve this is with Secure Shell (SSH) keys. This method will create both a public key (the keyhole) and a private key (your key for it). To create these, run the Unix commands below. 

###**Generate a private SSH key.**

Just like with physical keys, it is good practice to label them in case you have a lot of them. Here, you can label your keys with a comment using the `-C` option. It is standard to label keys by email address to easily distinguish who you will give access to your GitHub repos.

```python
%%bash
ssh-keygen -t ed25519 -C '[YOUR_EMAIL_HERE]' -f ~/.ssh/id_ed25519 -P ''
cat > ~/.ssh/config
Host *
  AddKeysToAgent yes
  IdentityFile ~/.ssh/id_ed25519
```

```python
%%bash
eval "$(ssh-agent -s)"
ssh-add ~/.ssh/id_ed25519
```

###**Generate and register a "keyhole" (public key).**

```python
%%bash
cat ~/.ssh/id_ed25519.pub
```

Next, add this public SSH key to your GitHub account here:
https://github.com/settings/keys

Once at this page, click on the green button circled in red below:

image.png

Then, give your new keyhole a name under **Title**, copy the text output the last code block into **Key**, and click the `Add SSH key` button to finialize your keyhole.  

image.png

Finally, you can verify that your key-keyhole pair works by running the command below. 

```python
%%bash
ssh -o "StrictHostKeyChecking no" -T git@github.com
```

If it does work, you will receive a message that includes the phrase `You've successfully authenticated`.

Note that whenever you restart your notebook, you will have to generate a new key and register a new keyhole on GitHub. As such, be sure to delete old keyholes from the https://github.com/settings/keys page to prevent your account from accumulating unnecessary keyholes like follows:

image.png
