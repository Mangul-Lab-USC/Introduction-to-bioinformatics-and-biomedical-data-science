##**Operators**

In addition to the assignment operator `=`, you may manipulate data types further with the following Python operators.

**Arithmetic operators ( +, -, /, \*, **)**

Below, these operators are used to calculate a patient's Body Mass Index (BMI).

````python
weightKG = 70 
heightCM = 180

heightM = heightCM * 0.01

BMI = weightKG / (heightM**2)
BMI
````

**Combine strings with with +**

When `+` is applied to strings, it performs a different task than when used with numbers: it concatenates them.

````python
genus = 'Mus'
species = 'musculus'

fullName = genus + ' ' + species
fullName
````

**Repeat strings with *.**

Similarly, `*` also performs a different task when used with strings.

````python
DNAMotif = 'TTAGGG'
telomereMinisatellite = DNAMotif * 25
telomereMinisatellite
````

**Test equivalency with == and !=.**

Variables can be compared using the values-are-equivalent operator `==` and values-are-NOT-equivalent operator `!=`.

````python
geneOneIsUpregulated = True
````

````python
geneOneIsUpregulated == False
````

````python
geneOneIsUpregulated != False
````

Furthermore, you may test more complex equivalencies with the addition of the logical operators AND (&/and) as well as OR (|/or).

````python
geneOneIsUpregulated = True
geneTwoIsUpregulated = False
````

````python
bothGenesAreUpregulated = (geneOneIsUpregulated == True) & (geneTwoIsUpregulated == True)
bothGenesAreUpregulated
````

````python
atLeastOneGeneIsUpregulated = (geneOneIsUpregulated == True) | (geneTwoIsUpregulated == True)
atLeastOneGeneIsUpregulated
````

Notice how, in this case, identical results can be obtained using the long form of both operators instead.

````python
bothGenesAreUpregulated = (geneOneIsUpregulated == True) and (geneTwoIsUpregulated == True)
bothGenesAreUpregulated
````

````python
atLeastOneGeneIsUpregulated = (geneOneIsUpregulated == True) or (geneTwoIsUpregulated == True)
atLeastOneGeneIsUpregulated
````

**Compare numbers with >, <, >=, and <=.**

The values of variables can also be compared in terms of relative "size". These operators are as follows:

*   `>` : greater than
*   `<` : less than
*   `>=` : greater than or equal to
*   `<=` : less than or equal to t



````python
42 > 0
````

````python
0.0 > 42.0
````

````python
42 >= 42
````

````python
42 <= 42
````

When these operators are used with string values, their values are instead compared lexicographically. 

````python
'a' < 'b' and 'b' < 'c'
````

````python
'a' > 'b'
````

##**Indexing and slicing**

In Python, the specific elements of ordered objects such as strings (of characters) and lists can accessed using indexing (with brackets []) and slicing (with brackets with one or more colons [:]).

**String indexing**

Each position in a string can be called by two different labels: a positive index and a negative index.

image.png

````python
mouseStrain = 'C57BL/6J-ApcMin/J'
mouseStrain[0]
````

````python
mouseStrain[-1]
````

**String slicing**

To access multiple positions in a string, they can called using the interval [start:stop-1]. 

image.png

In mathematics courses, you may have seen such intervals described as semi-open intervals. 

image.png

We can use string slicing to extract information from a lab mouse strain designations that follow the same format. 

````python
mouseStrain = 'C57BL/6J-ApcMin/J'
backgroundStrain = mouseStrain[0:7]
backgroundStrain
````

````python
mutation = mouseStrain[9:-2]
mutation
````

````python
mouseStrainTwo = 'C57BL/6J-Sting1gt/J'
mutationTwo = mouseStrainTwo[9:-2]
mutationTwo
````

**List indexing & slicing**

Like strings, lists also have indexes and can be sliced. 

image.png

````python
DNANucleotides = ['A', 'T', 'C', 'G']
DNANucleotides[2:4]
````

List indexing can also be used to replace elements in a list. 

````python
RNANucleotides = DNANucleotides
RNANucleotides[1] = 'U'
RNANucleotides
````

## **Dictionary usage**

Unlike strings and lists, dictionaries are unordered. So, instead of indexing and slicing, the elements of dictionaries must be manipulated in the following ways.

**Select an element of a dictionary using a key.**

Recall that dictionarys are unordered, therefore indexing & slicing will not work.

````python
AAToCodon = { 'Met':'AUG', 'Leu': ['UUA', 'UUG', 'CUU', 'CUC', 'CUA', 'CUG'] }
AAToCodon['Leu']
````

**Add a new element to the dictionary.**

If a key in a dictionary does not exist already, you can create a new key-value pair as follows:

````python
AAToCodon['Ser'] = ['UCU', 'UCC', 'UCA', 'UCG', 'AGU', 'AGC']
AAToCodon
````

**Overwrite an element in the dictionary.**

In yeast mitochondria, an alternative codon table is used in which AUA codes for methionine instead of isoleucine.

````python
AAToCodon['Met'] = ['AUG', 'AUA'] 
AAToCodon
````
